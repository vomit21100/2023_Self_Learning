```python
import pandas as pd
from sklearn import datasets
```

```python
data = datasets.load_iris()
iris = pd.DataFrame(data['data'], columns = data['feature_names'])
iris.loc[:, 'label'] = data['target']
iris.head()
```

```python
iris.describe()
```

## 計算各屬性Outliners的數量

```python
q1 = iris.quantile(.25)
q3 = iris.quantile(.75)
IQR = q3 - q1
outliers = iris[((iris < (q1 - 1.5*IQR)) | (iris > (q3 + 1.5*IQR)))]
outliers.count()
```

```python
outliers.dropna(how='all', inplace=True)
outliers
```

## 替換數值

```python
outliers.loc[outliers.index, 'sepal width (cm)'] = iris['sepal width (cm)'].median()
outliers.loc[outliers.index]
```
